<b>5. What do the keyword arguments delimiter and line terminator do?

delimiter specifies the separator like comma. While, lineterminator specified the end of line character like \n

```python
# Example of csv.writer
import csv

with open('sample.csv', 'w') as f:
    csv_writer = csv.writer(f, delimiter=',', lineterminator='\n')
    csv_writer.writerows([['a', 'b', 'c', 'd'], ['e', 'f', 'g', 'h']])
```

```python
with open('sample.csv', 'r') as f:
    csv_reader = csv.reader(f)
    for row in csv_reader:
        print(row)
```

<b>6. What function takes a string of JSON data and returns a Python data structure?

json.loads()

```python
import json

s = '{"a": [1, 2, 3]}'
d = json.loads(s)
print(d)
print(type(d))
```

<b>7. What function takes a Python data structure and returns a string of JSON data?

json.dumps()

```python
d = {"a": [1, 2, 3]}
s = json.dumps(d)
print(s)
print(type(s))
```
